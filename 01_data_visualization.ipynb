## CIFAR-10

The primary dataset used for testing this network is CIFAR-10. CIFAR-10 is a standard benchmark dataset for image classification. It consists of 60,000 32x32x3 (RGB) images. This yields a feature vector of 3,072. The dataset may be downloaded [here](https://www.cs.toronto.edu/~kriz/cifar.html) but also comes standard with Keras.

The 10 classes in CIFAR-10:
- airplanes
- automobiles
- birds
- cats
- deer
- dogs
- frogs
- horses
- ships
- trucks

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

There are 50,000 images in the train, 10,000 in the test.  
The images are 32x32 pixels, and have three channels: RGB.

```python
# Define a function to help visualize CIFAR-10.
def display_cifar(x_train, y_train, num, ncols, figsize):
    """
    (pandas df, series, int, int, tuple) -> grid of training images with labels
    
    Returns a plot of of images from the CIFAR-10 dataset and their labels.    
    """   
    
    # Save the CIFAR-10 labels as a list inside the function.
    cifar_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 
    'frog', 'horse', 'ship', 'truck']
    
    # Divide the numbers of images to display by the number of columns
    # to make sure you have enough rows.
    nrows = int(np.ceil(num/ncols)) 
    
    fig = plt.figure(figsize=figsize)

    # Add a centered title to the figure.
    plt.suptitle('CIFAR-10 Images and their Labels', fontsize=14)
    
    # Adjust the height for space between subplots.
    plt.subplots_adjust(hspace=.5)
    
    # Iterate through the number of images to display.
    for i in range(num):

        # Add a subplot in the appropriate position.
        fig.add_subplot(nrows, ncols, i + 1)
        
        # Plot the image at that index.
        plt.imshow(X_train[i], interpolation='lanczos')
        
        # Set the label for each image.
        plt.title(cifar_classes[y_train[i][0]])

    return
```

```python
# Display the first 18 images of CIFAR-10 and their labels.
display_cifar(X_train, y_train, 18, 6, (12,6))
```

```python
# Display the first 8 images of CIFAR-10 and their labels.
display_cifar(X_train, y_train, 8, 4, (10,5))
```
